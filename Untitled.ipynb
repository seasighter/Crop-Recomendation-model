```python
# creating new dataframe without label
z=data.drop(['label'],axis=1)     # dropping label columns
z=data.loc[:,z.columns].values
x_df=pd.DataFrame(z)
x_df.head()
```

```python
# we will  use elbow mthod to find clusters in clustering
from sklearn.cluster import KMeans
plt.rcParams['figure.figsize']=(10,4)
wcss=[]
for i in range(1,11):
    km = KMeans(n_clusters =i, init= 'k-means++', max_iter=300, n_init=10, random_state=0)
    km.fit(z)
    wcss.append(km.inertia_) # here km.inertia will give sum of sqaure error
#plot the results 
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel('No.of cluster')
plt.ylabel('wcss')
plt.show()
    
```

Observation:

we get two elbow at 3 and 4 .As per elbow method definition we take the last one so our no. of cluster will be # 4

```python
km=KMeans(n_clusters=4,max_iter=300,n_init=10,random_state=0)
y_means=km.fit_predict(z)
a=data['label']
y_means=pd.DataFrame(y_means)
w=pd.concat([y_means,a],axis=1)
w=w.rename(columns={0:'cluster'})   # renaming column name from 0 to cluster

```

```python
#check  cluster of each group
for i in range(0,4):
    print("Crops is cluster",i,w[w['cluster']==i]['label'].unique())
    print("------------------------------------------------")
                                                 
```

Observation :
    
    some of the crops are more than in one clusters .This is known as soft clustering
    as we can see that is in our case also as papaya occurs in both cluster 0 and 3
    In hard clustering we can able to see data  points are completely in one cluster or not

```python
for i in range(0,4):
    counts=w[w['cluster']==i]['label'].value_counts()
    d = w.loc[w['label'].isin(counts.index[counts>=50])]
    d=d['label'].value_counts()
    print('Crops in Cluster', i,':', list(d.index))
    print('--------------------------------------------------------------------------------')
```

```python
x=data.drop(["label"],axis=1)
x.head()
```

```python
y=data["label"]
y.shape
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)  # ramdom state =0 means shuffling is off in every execution we get same result 
```

```python
len(x_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf_knn=KNeighborsClassifier(n_neighbors=3)
clf_knn.fit(x_train,y_train)
y1_pred1=clf_knn.predict(x_test)
print("Accuracy Score in KNN:",accuracy_score(y_test,y1_pred1))

from sklearn.svm import SVC
clf_svc=SVC()
clf_svc.fit(x_train,y_train)
y2_pred2=clf_svc.predict(x_test)
print("Accuracy Score in SVC:",accuracy_score(y_test,y2_pred2))

from sklearn.tree import DecisionTreeClassifier
clf_dtc=DecisionTreeClassifier()
clf_dtc.fit(x_train,y_train)
y3_pred3=clf_dtc.predict(x_test)
print("Accuracy Score in Decision tree:",accuracy_score(y_test,y3_pred3))

from sklearn.ensemble import RandomForestClassifier
clf_rdc=RandomForestClassifier(random_state=1)
clf_rdc.fit(x_train,y_train)
y4_pred4=clf_rdc.predict(x_test)
print("Accuracy Score in Random forest:",accuracy_score(y_test,y4_pred4))
```

Random forest gave us the best result

```python
y_train_pred=clf_rdc.predict(x_train)
print("Accuracy Score of random forest: ",accuracy_score(y_train,y_train_pred))
```
